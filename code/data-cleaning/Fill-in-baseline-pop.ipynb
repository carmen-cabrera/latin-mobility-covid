```python
### CONSIDER: DO YOU WANT TO WORK WITH FILES LOCALLY (WD) OR REMOTELY (DIRECTORY)
### movement files will always be in "directory"

filename = wd + '/baseline/counter_pop.txt'
with open(filename, 'r') as file:
    start_i = file.read()
print(start_i)

df_bl = pd.read_csv(wd + '/baseline/baseline_pop.csv')


for i in range(int(start_i), len(files)):
    
    if i%10 == 0:
        print(i/len(files)*100)
    
    df_bl = pd.read_csv(wd + '/baseline/baseline_pop.csv')

    file = files[i]
    df_pops = pd.read_csv(wd + '/pop/' + file, index_col=0)

    df_pops_bl = df_pops#.dropna(subset=['n_baseline']).reset_index(drop=True)
    df_pops_bl = df_pops_bl.dropna(subset=['FID']).reset_index(drop=True)
    wday = datetime.strptime(file[-19:-9], "%Y-%m-%d").weekday()

    for j in range(len(df_pops_bl)):

        ID = int(df_pops_bl.loc[j, 'FID'])

        df_bl_t = df_bl[df_bl['FID']==ID]
        value = df_pops_bl.loc[j, 'n_baseline']

        if len(df_bl_t) == 0:
            row = [int(ID)] + [-999 if k != wday else value for k in range(7)]
            df_bl.loc[len(df_bl)] = row

        else:
            if len(df_bl_t.index) == 1:
                if (df_bl.loc[df_bl_t.index[0], str(wday)] > 0) == False:
                    df_bl.loc[df_bl_t.index[0], str(wday)] = value
            else:
                print('more than one rows in df_bl_t')
            
    filename = wd + '/baseline/counter_pop.txt'
    with open(filename, 'w') as file:
        file.write(str(i))
        
    df_bl.to_csv(wd + '/baseline/baseline_pop.csv', index=False)
```
