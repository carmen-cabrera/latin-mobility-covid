# Imports

```python
import os
import pandas as pd
import geopandas as gpd
import numpy as np


def init_counts():
    return {
        'total_records': 0,
        'missing_records': 0,
        'missing_forced_zero_no_builtup': 0,
        'model_imputed_records': 0,
    }


def update_counts(counts, series, eligible_mask):
    missing_mask = ~(series > 0)
    forced_zero_mask = missing_mask & ~eligible_mask
    imputed_mask = missing_mask & eligible_mask

    counts['total_records'] += int(len(series))
    counts['missing_records'] += int(missing_mask.sum())
    counts['missing_forced_zero_no_builtup'] += int(forced_zero_mask.sum())
    counts['model_imputed_records'] += int(imputed_mask.sum())


def print_counts(counts):
    summary = {
        'total_records': counts['total_records'],
        'missing_records': counts['missing_records'],
        'pct_total_imputed_excluding_no_builtup': (
            counts['model_imputed_records'] / counts['total_records'] * 100
            if counts['total_records'] > 0 else float('nan')
        )
    }
    print(summary)
```

# Define country and parameters

```python
# Select target country
country = 'Colombia'

# Set country-specific parameters: ISO codes and buffer size (in meters)
if country == 'Argentina':
    country_short = 'ARG'   # ISO 3-letter code
    country_code = 'AR'     # ISO 2-letter code
elif country == 'Chile':
    country_short = 'CHL'
    country_code = 'CL'
elif country == 'Colombia':
    country_short = 'COL'
    country_code = 'CO'
# Uncomment the following if Mexico is to be included in the analysis
# elif country == 'Mexico':
#     country_short = 'MEX'
#     country_code = 'MX'
```

# Baseline Facebook population

```python
pop_meta = gpd.read_file(
    os.path.join(
        wd, 'data', 'outputs', country_short, 'grids-with-data',
        'popcell-baseline-imput-pop-builtup', 'popcell-baseline-imput-pop-builtup.gpkg'
    )
)
if 'FID' not in pop_meta.columns:
    pop_meta['FID'] = pop_meta.index
pop_meta = pop_meta[['FID', 'has_builtup']]

df_bl = pd.read_csv(
    os.path.join(wd, 'data', 'outputs', country_short, 'baseline', 'baseline_pop.csv')
)
df_bl = df_bl.merge(pop_meta, on='FID', how='left')
df_bl['has_builtup'] = df_bl['has_builtup'].fillna(0).astype(int)

counts = init_counts()

for i in range(6):
    col = str(i)
    eligible_mask = df_bl['has_builtup'] == 1
    update_counts(counts, df_bl[col], eligible_mask)

print_counts(counts)
```

# Baseline Facebook movement

```python
df_bl = pd.read_csv(
    os.path.join(
        wd, 'data', 'outputs', country_short, 'baseline',
        'movcell-baseline-imput-mov-dist-with-exo-var-builtup-long.csv'
    )
)

pair_summary = (
    df_bl.groupby(['O', 'D'], as_index=False)
    .agg(
        any_missing=('was_missing', 'max'),
        all_missing=('was_missing', 'min'),
        any_model_imputed=('model_imputed', 'max')
    )
)

summary = {
    'total_records': int(len(pair_summary)),
    'missing_records': int(pair_summary['any_missing'].sum()),
    'pct_total_imputed_excluding_no_builtup': float((pair_summary['all_missing'] & pair_summary['any_model_imputed']).mean() * 100)
}
print(summary)
```

# Crisis Facebook population

```python
pop_meta = gpd.read_file(
    os.path.join(
        wd, 'data', 'outputs', country_short, 'grids-with-data',
        'popcell-baseline-imput-pop-builtup', 'popcell-baseline-imput-pop-builtup.gpkg'
    )
)
if 'FID' not in pop_meta.columns:
    pop_meta['FID'] = pop_meta.index
pop_meta = pop_meta[['FID', 'has_builtup']]

file_path = os.path.join(
    wd, 'data', 'outputs', country_short, 'evo', 'pop_evo_popcell_raw.csv'
)
df_cr = pd.read_csv(file_path).drop(columns=['Unnamed: 0'], errors='ignore')
id_col = 'FID' if 'FID' in df_cr.columns else df_cr.columns[0]
df_cr = df_cr.merge(pop_meta, left_on=id_col, right_on='FID', how='left')
df_cr['has_builtup'] = df_cr['has_builtup'].fillna(0).astype(int)

data_cols = [c for c in df_cr.columns if c not in {id_col, 'FID', 'has_builtup'}]
counts = init_counts()

for column in data_cols:
    eligible_mask = df_cr['has_builtup'] == 1
    update_counts(counts, df_cr[column], eligible_mask)

print_counts(counts)
```

# Crisis Facebook movement

```python
file_path = os.path.join(
    wd, 'data', 'outputs', country_short, 'evo', 'mov_evo_dist_raw.csv'
)
df_cr = pd.read_csv(file_path).drop(columns=['Unnamed: 0'], errors='ignore')

mov_meta = gpd.read_file(
    os.path.join(
        wd, 'data', 'outputs', country_short, 'grids-with-data',
        'movcell-baseline-imput-pop-with-exo-var-builtup',
        'movcell-baseline-imput-pop-with-exo-var-builtup.gpkg'
    )
)
if 'FID' not in mov_meta.columns:
    mov_meta['FID'] = mov_meta.index
builtup_lookup = mov_meta.set_index('FID')['has_builtup'].fillna(0).astype(int)

data_cols = list(df_cr.columns[2:])
eligible_od = df_cr['O'].map(builtup_lookup).fillna(0).eq(1) & df_cr['D'].map(builtup_lookup).fillna(0).eq(1)
any_positive = (df_cr[data_cols] > 0).any(axis=1)
any_missing = ~(df_cr[data_cols] > 0).all(axis=1)
model_imputed_pair = eligible_od & (~any_positive)

summary = {
    'total_records': int(len(df_cr)),
    'missing_records': int(any_missing.sum()),
    'pct_total_imputed_excluding_no_builtup': float(model_imputed_pair.mean() * 100)
}
print(summary)
```
