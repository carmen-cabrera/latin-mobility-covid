# Helper functions

```python
def read_gpkg_attributes(path):
    con = sqlite3.connect(path)
    try:
        table_name = con.execute('SELECT table_name FROM gpkg_contents LIMIT 1').fetchone()[0]
        df = pd.read_sql_query(f'SELECT * FROM "{table_name}"', con)
    finally:
        con.close()
    return df


def normalise_id_column(df, preferred_name='FID'):
    df = df.copy()
    if preferred_name not in df.columns and preferred_name.lower() in df.columns:
        df = df.rename(columns={preferred_name.lower(): preferred_name})
    return df


def summarise_imputation(raw_df, imputed_df, value_cols, id_cols=None):
    raw_values = raw_df[value_cols].apply(pd.to_numeric, errors='coerce')
    imputed_values = imputed_df[value_cols].apply(pd.to_numeric, errors='coerce')

    missing_before_mask = raw_values.isna() | raw_values.isin([-1, -999])
    missing_after_mask = imputed_values.isna() | imputed_values.isin([-1, -999])

    total_records = int(raw_values.size)
    records_missing_before = int(missing_before_mask.sum().sum())
    records_missing_after = int(missing_after_mask.sum().sum())
    imputed_records = max(records_missing_before - records_missing_after, 0)

    observed_total_raw = raw_values.where(~missing_before_mask, 0).sum().sum()
    final_total = imputed_values.where(~missing_after_mask, 0).sum().sum()
    imputed_total = max(final_total - observed_total_raw, 0)

    summary = {
        'total_records': total_records,
        'records_missing_before': records_missing_before,
        'records_missing_after': records_missing_after,
        'imputed_records': int(imputed_records),
        'pct_records_imputed': float(imputed_records / total_records * 100) if total_records else 0.0,
        'observed_total_raw': float(observed_total_raw),
        'imputed_total': float(imputed_total),
        'final_total': float(final_total),
        'pct_total_imputed': float(imputed_total / final_total * 100) if final_total else 0.0,
    }
    return pd.Series(summary)


def weekday_cols(df):
    return [str(i) for i in range(7) if str(i) in df.columns]


def date_value_cols(df, id_cols):
    return [col for col in df.columns if col not in id_cols and not str(col).startswith('Unnamed')]
```

```python
def format_summary(summary):
    summary = summary.copy()

    integer_fields = [
        'total_records',
        'records_missing_before',
        'records_missing_after',
        'imputed_records',
    ]
    total_fields = ['observed_total_raw', 'imputed_total', 'final_total']
    percentage_fields = ['pct_records_imputed', 'pct_total_imputed']

    for field in integer_fields:
        summary[field] = f"{int(round(summary[field])):,}"

    for field in total_fields:
        summary[field] = f"{summary[field]:,.2f}"

    for field in percentage_fields:
        summary[field] = f"{summary[field]:,.2f}%"

    return summary
```

# Baseline Facebook population

```python
baseline_pop_raw = pd.read_csv(
    os.path.join(wd, 'data', 'outputs', country_short, 'baseline', 'baseline_pop.csv')
)
baseline_pop_imputed = normalise_id_column(
    read_gpkg_attributes(
        os.path.join(
            wd,
            'data',
            'outputs',
            country_short,
            'grids-with-data',
            'popcell-baseline-imput-pop',
            'popcell-baseline-imput-pop.gpkg',
        )
    )
)

summary_baseline_population = summarise_imputation(
    normalise_id_column(baseline_pop_raw),
    baseline_pop_imputed,
    value_cols=weekday_cols(baseline_pop_raw),
    id_cols=['FID'],
)
format_summary(summary_baseline_population)
```

# Baseline Facebook movement

```python
baseline_mov_raw = pd.read_csv(
    os.path.join(wd, 'data', 'outputs', country_short, 'baseline', 'baseline_mov.csv')
)
baseline_mov_raw = baseline_mov_raw.fillna(-1).replace(-999, -1)
baseline_mov_raw = baseline_mov_raw[baseline_mov_raw['O'] != baseline_mov_raw['D']].reset_index(drop=True)

baseline_mov_imputed = pd.read_csv(
    os.path.join(
        wd,
        'data',
        'outputs',
        country_short,
        'baseline',
        'movcell-baseline-imput-mov-dist-with-exo-var-flatten-censored10.csv',
    )
).drop(columns=['Unnamed: 0'], errors='ignore')

summary_baseline_movement = summarise_imputation(
    baseline_mov_raw,
    baseline_mov_imputed,
    value_cols=weekday_cols(baseline_mov_raw),
    id_cols=['O', 'D'],
)
format_summary(summary_baseline_movement)
```

# Crisis Facebook population

```python
crisis_pop_raw = pd.read_csv(
    os.path.join(wd, 'data', 'outputs', country_short, 'evo', 'pop_evo_popcell_raw.csv')
).drop(columns=['Unnamed: 0'], errors='ignore')
crisis_pop_imputed = pd.read_csv(
    os.path.join(wd, 'data', 'outputs', country_short, 'evo', 'pop_evo_popcell.csv')
).drop(columns=['Unnamed: 0'], errors='ignore')

summary_crisis_population = summarise_imputation(
    crisis_pop_raw,
    crisis_pop_imputed,
    value_cols=date_value_cols(crisis_pop_raw, ['FID']),
    id_cols=['FID'],
)
format_summary(summary_crisis_population)
```

# Crisis Facebook movement

```python
crisis_mov_raw = pd.read_csv(
    os.path.join(wd, 'data', 'outputs', country_short, 'evo', 'mov_evo_dist_raw.csv')
).drop(columns=['Unnamed: 0'], errors='ignore')
crisis_mov_imputed = pd.read_csv(
    os.path.join(wd, 'data', 'outputs', country_short, 'evo', 'mov_evo_dist_censored10.csv')
).drop(columns=['Unnamed: 0'], errors='ignore')

summary_crisis_movement = summarise_imputation(
    crisis_mov_raw,
    crisis_mov_imputed,
    value_cols=date_value_cols(crisis_mov_raw, ['O', 'D']),
    id_cols=['O', 'D'],
)
format_summary(summary_crisis_movement)
```
