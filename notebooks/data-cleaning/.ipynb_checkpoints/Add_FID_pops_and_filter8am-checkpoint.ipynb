## Add IDs to population data and save

```python
directory = '/Volumes/RECAST/data/populations/facebook/' + country
files = sorted(os.listdir(directory))
files = [file for file in files if '_0800.csv' in file]
```

```python
for i in range(len(files)):   
    
    if i%50 == 0:
        print(i/len(files)*100)
    
    try:
        
        file = files[i]

        df_pops = pd.read_csv(directory + '/' + file)
        df_pops = df_pops[df_pops['country']==country_code].reset_index(drop=True)
        
        index_to_drop = []
        df_pops['FID'] = np.ones(len(df_pops))*(-1)
        
        for j in range(len(df_pops)):
            try:
                point = gpd.GeoDataFrame({'geometry': [Point(df_pops.loc[j, 'lon'], df_pops.loc[j, 'lat'])]}).set_crs('EPSG:4326')
                point = point.to_crs('EPSG:3857').buffer(buffer).to_crs('EPSG:4326')
                overlap_point = point[0].intersects(grid_pop['geometry'])
                df_pops.loc[j, 'FID'] = grid_pop.loc[np.where(overlap_point == True)[0][0]]['FID']
            except:
                index_to_drop.append(j)
        
        df_pops = df_pops.drop(index_to_drop).reset_index(drop=True)

        df_pops.to_csv(wd + '/data/outputs/' + country_short + '/pop/' + file)
    
    except:
        print(i)
    
#         wd + '/data/outputs/' + country_short + '/pop/' + file)
```
